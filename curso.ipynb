## 🧠 Concepto Fundamental: ¿Qué es un Script?
Un script es simplemente una **lista de mandados** para la computadora. En lugar de escribir 5 comandos uno tras otro cada vez, los guardas en un archivo y le pides a la terminal que los ejecute todos de un tirón.

---

## 🛠️ Clase 1: El ADN del Script (`01-hola.sh`)

Crea tu primer archivo en VS Code y escribe esto. Pero antes de escribirlo, **entiende qué hace cada parte**:

```bash
#!/bin/bash

# Este es un comentario. El robot lo ignora, pero nosotros no.
echo "-------------------------------------"
echo "INICIANDO SISTEMA DE AUTOMATIZACIÓN"
echo "-------------------------------------"

# Imprimir un mensaje en pantalla
echo "Hola, Pablo. El entorno está listo."
```

### 🗝️ Lo que debes aprender hoy:
1.  **El Shebang (`#!/bin/bash`):** Es la línea más importante. Le dice a la computadora: *"¡Ojo! No leas esto como texto común, usa el intérprete de Bash para ejecutar estas órdenes"*. Sin esto, el script es solo una carta sin dirección.
2.  **`echo`:** Es el comando para "hablar". Todo lo que pongas después de `echo` saldrá impreso en la terminal.
3.  **Extensiones:** Aunque en Linux la extensión no es obligatoria, usamos `.sh` para que nosotros (y VS Code) sepamos que es un script de Shell.

## 🗣️ Clase 2: Memoria e Interacción (`02-interactivo.sh`)

Aquí aprendemos a que el script no sea un monólogo, sino un diálogo.

```bash
#!/bin/bash

# Declarar una variable fija (Caja con etiqueta)
PROFESOR="Pol Monsalvo"

echo "Bienvenido a la clase de $PROFESOR"

# Pedir datos al usuario (La terminal se detiene y espera)
echo "¿En qué proyecto vamos a trabajar hoy?"
read PROYECTO

echo "Perfecto, preparando el terreno para: $PROYECTO..."
```

### 🗝️ Conceptos de "Programador Pro":
* **Variables:** Son "cajas" donde guardamos información. Para crearlas, se usa `NOMBRE="Valor"` (sin espacios alrededor del igual). Para usarlas, se les pone el signo de dólar delante: `$NOMBRE`.
* **`read`:** Es el comando que pausa el script. Es fundamental para crear herramientas que otros puedan usar.

## 🚀 Cómo "Encender" tus scripts (Paso Crítico)

Para que tú aprendas a solucionar problemas, debes conocer estos dos errores típicos:

1.  **Permiso Denegado:** Si intentas correr el script y te dice que no tienes permiso, usa el comando:

    *(El `./` le dice a la terminal: "Busca el archivo AQUÍ, en esta carpeta").*

### 📄 Archivo: `02-interactivo.sh`

```bash
#!/bin/bash

# 1. Pedimos el nombre
echo "Hola, soy tu asistente de terminal. ¿Cómo te llamas?"
read NOMBRE

# 2. Pedimos el año
echo "Mucho gusto, $NOMBRE. ¿En qué año estamos?"
read ANIO

# 3. Devolvemos el saludo personalizado
echo "------------------------------------------------"
echo "¡Perfecto! Hola $NOMBRE, estamos en el año $ANIO."
echo "Es un gran momento para seguir programando."
echo "------------------------------------------------"
```

---

### 🧠 ¿Qué estamos aprendiendo aquí? (Análisis de Mentor)

1.  **El espacio importa:** Fíjate que en la línea `read NOMBRE`, la terminal se queda parpadeando. El usuario escribe algo, aprieta **Enter**, y en ese instante, el texto "viaja" dentro de la palabra `NOMBRE`.
2.  **Llamar a la variable:** Cuando escribes `echo "Mucho gusto, $NOMBRE"`, el signo **`$`** actúa como una llave. Si no pones el `$`, la terminal escribirá literalmente la palabra "NOMBRE" en lugar del nombre del alumno.
3.  **Memoria temporal:** Estas variables solo viven mientras el script está corriendo. Una vez que termina, la computadora "olvida" esos datos.
