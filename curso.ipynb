## 🧠 Concepto Fundamental: ¿Qué es un Script?
Un script es simplemente una **lista de mandados** para la computadora. En lugar de escribir 5 comandos uno tras otro cada vez, los guardas en un archivo y le pides a la terminal que los ejecute todos de un tirón.

Crea tu primer archivo en VS Code y escribe esto. Pero antes de escribirlo, **entiende qué hace cada parte**:

```bash
#!/bin/bash

# Este es un comentario. El robot lo ignora, pero nosotros no.
echo "-------------------------------------"
echo "INICIANDO SISTEMA DE AUTOMATIZACIÓN"
echo "-------------------------------------"

# Imprimir un mensaje en pantalla
echo "Hola, Pablo. El entorno está listo."
```

### 🗝️ Lo que debes aprender hoy:
1.  **El Shebang (`#!/bin/bash`):** Es la línea más importante. Le dice a la computadora: *"¡Ojo! No leas esto como texto común, usa el intérprete de Bash para ejecutar estas órdenes"*. Sin esto, el script es solo una carta sin dirección.
2.  **`echo`:** Es el comando para "hablar". Todo lo que pongas después de `echo` saldrá impreso en la terminal.
3.  **Extensiones:** Aunque en Linux la extensión no es obligatoria, usamos `.sh` para que nosotros (y VS Code) sepamos que es un script de Shell.

* **`read`:** Es el comando que pausa el script. Es fundamental para crear herramientas que otros puedan usar.

2.  **La Sintaxis Matemática (`$((...))`):** En Bash, si pones `EDAD + 1`, la terminal cree que es texto. Para que entienda que es una **operación matemática**, debes envolverlo siempre en doble paréntesis y con el signo pesos: `$(( cuenta ))`.
    
3.  **La Salida (`echo`):** Nota cómo mezclamos texto fijo (lo que va entre comillas) con el valor de las variables (lo que tiene el `$`).

## 🛠️ Desafío Extra: El Calculador de Edad Crónico
Crea un nuevo archivo llamado `02-desafio-calculadora.sh`. Aquí está la estructura para que la repliques y comprendas la lógica.

### 📄 Archivo: `02-desafio-calculadora.sh`

```bash
#!/bin/bash

# --- 1. CAPTURA DE DATOS ---
echo "--- CALCULADORA DE EDAD AUTOMÁTICA ---"

read -p "¿Cómo te llamas? " NOMBRE
read -p "Hola $NOMBRE, ¿en qué año estamos? " ANIO_ACTUAL
read -p "¿En qué año naciste? " ANIO_NACIMIENTO

# --- 2. PROCESAMIENTO (LA MAGIA) ---
# Restamos el año actual menos el año de nacimiento
EDAD_CALCULADA=$((ANIO_ACTUAL - ANIO_NACIMIENTO))

# --- 3. SALIDA ---
echo "------------------------------------------------"
echo "Análisis completado para $NOMBRE:"
echo "Si naciste en $ANIO_NACIMIENTO y estamos en $ANIO_ACTUAL..."
echo "¡Entonces tienes (o cumplirás) $EDAD_CALCULADA años!"
echo "------------------------------------------------"
```

---

### 🧠 Lección de Mentor: ¿Qué hay de nuevo aquí?

1.  **`read -p` (Input más limpio):** Fíjate que usé `read -p "Mensaje" VARIABLE`. El parámetro `-p` significa "prompt". Sirve para poner la pregunta y la entrada de texto en la **misma línea**. Hace que el script se vea mucho más profesional y ordenado.
    

2.  **La Resta en Bash:**
    La sintaxis sigue siendo `$(( ... ))`. No importa si sumas, restas o multiplicas (usando `*`), siempre debes usar los paréntesis dobles.

## 🏗️ Módulo 2: El Arquitecto - Estructura Automática

El objetivo de esta clase es aprender a usar la terminal para organizar un proyecto web en un segundo. Pero hoy vamos a subir la apuesta: usaremos una **variable** para que el usuario decida el nombre de la carpeta del proyecto.

### 📄 Archivo: `03-constructor-web.sh`

Copia esto en tu VS Code, analicemos la lógica y luego documéntalo en tu notebook.

```bash
#!/bin/bash

echo "--- 🛠️ GENERADOR DE ESTRUCTURA WEB 1.0 ---"

# 1. Preguntamos el nombre del proyecto
read -p "Nombre de la carpeta del proyecto: " PROYECTO

# 2. Creamos la carpeta principal
mkdir "$PROYECTO"
echo "✅ Carpeta '$PROYECTO' creada."

# 3. Entramos a la carpeta y creamos las subcarpetas
cd "$PROYECTO"
mkdir css js img video audio
echo "✅ Subcarpetas (css, js, img, video, audio) creadas."

# 4. Creamos los archivos base vacíos
touch index.html css/style.css js/script.js
echo "✅ Archivos base (index, style, script) creados."

echo "------------------------------------------------"
echo "🚀 ¡Todo listo! Tu proyecto $PROYECTO está preparado."
echo "Escribe: cd $PROYECTO y luego: code ."
echo "------------------------------------------------"
```

---

### 🧠 Lección de Mentor: Los nuevos comandos

En este script estamos usando "Verbos de Acción" de Linux:

1.  **`mkdir` (Make Directory):** Crea una carpeta. Nota que puse `"$PROYECTO"` entre comillas; esto es para que si el nombre tiene espacios, Linux no se confunda.
2.  **`cd` (Change Directory):** "Entra" en la carpeta. Es fundamental porque si no entramos, las subcarpetas se crearían afuera.
3.  **`touch`:** Es el comando para crear un archivo vacío. Fíjate que puedes crear varios a la vez: `touch uno.html dos.css`.
4.  **Rutas Relativas:** Mira la línea `touch css/style.css`. Le estamos diciendo: *"Crea el archivo style.css DENTRO de la carpeta css"*.



---

### ✍️ Nota para tu Cuaderno `.ipynb`

> **Concepto de "Directorio Actual":**
> Al programar scripts que crean carpetas, siempre debemos ser conscientes de dónde estamos parados (`pwd`). El comando `cd` es nuestro volante para movernos por el disco duro desde el script.

### 🧠 Lección de Mentor: Los Operadores `>` y `>>`

En la terminal, puedes enviar el resultado de un comando hacia un archivo en lugar de hacia la pantalla:
1.  **`>` (Sobrescribir):** Borra todo lo que haya en el archivo y pone el texto nuevo. (Como pintar una pared de cero).
2.  **`>>` (Añadir):** Pone el texto al final de lo que ya existe. (Como pegar un sticker en una pared que ya tiene cuadros).

## 🧠 Módulo 3: El Script Inteligente (Condicionales)

Vamos a crear un archivo que solucione ese desorden. El objetivo es que el script verifique si ya existe una carpeta con ese nombre. Si existe, nos dará un aviso y **no hará nada**.

### 📄 Archivo: `04-validador-proyectos.sh`

```bash
#!/bin/bash

echo "--- 🛡️ CONSTRUCTOR CON SEGURIDAD ---"

read -p "Nombre del proyecto: " PROYECTO

# --- EL CORAZÓN DEL SCRIPT: EL IF ---
# -d significa "directory". Preguntamos: ¿Existe este directorio?

if [ -d "$PROYECTO" ]; then
    echo "-------------------------------------------------------"
    echo "⚠️ ERROR: La carpeta '$PROYECTO' ya existe."
    echo "Borra la carpeta o elige otro nombre para evitar líos."
    echo "-------------------------------------------------------"
else
    echo "✅ Nombre disponible. Iniciando construcción..."
    
    mkdir "$PROYECTO"
    cd "$PROYECTO"
    mkdir css js img
    touch index.html css/style.css js/script.js
    
    echo "🚀 Proyecto '$PROYECTO' creado con éxito."
fi
```

---

### 🧠 Lección de Mentor: La anatomía del `if`

Para tu cuaderno `.ipynb`, esta estructura es vital. Es el primer paso hacia la **IA lógica** en Bash.

1.  **`if [ -d "$VARIABLE" ]; then`**: 
    * `if`: Si...
    * `[ ]`: Los corchetes son como una balanza. Adentro se hace la comparación. **Ojo:** Siempre debe haber un espacio después del primer corchete y antes del último.
    * `-d`: Es un "test" para carpetas (directorios).
    * `then`: Entonces...
2.  **`else`**: Si la condición de arriba no se cumple (la carpeta NO existe), haz esto otro.
3.  **`fi`**: Es "if" al revés. Sirve para cerrar la estructura. En Bash, todo lo que se abre se cierra al revés.

# 🌀 Módulo 4: El Poder de la Repetición (Bucles `for`)

Imagínate que tienes 30 alumnos y necesitas crear una carpeta para cada uno, o que necesitas crear 12 carpetas para los 12 meses del año. Hacerlo a mano es aburrido y propenso a errores. Aquí es donde entra el **Bucle `for`**.

El `for` le dice a la computadora: *"Para cada elemento de esta lista, repite esta acción"*.

### 📄 Archivo: `07-bucle-carpetas.sh`

Crea este archivo y mira qué elegancia de código:

```bash
#!/bin/bash

echo "--- 📂 GENERADOR DE CARPETAS EN MASA ---"

# El bucle irá del 1 al 5
for i in {1..5}
do
   echo "Creando la carpeta: Clase-$i"
   mkdir "Clase-$i"
   
   # Opcional: Crear un archivo de notas dentro de cada una
   touch "Clase-$i/notas.txt"
done

echo "----------------------------------------"
echo "✅ ¡Se crearon 5 carpetas y 5 archivos en 1 segundo!"
```

---

### 🧠 Lección de Mentor: La Anatomía del Bucle

Para tu cuaderno `.ipynb`, esta estructura es oro puro:

1.  **`for i in {1..5}`**: 
    * `i`: Es una variable temporal (puedes llamarla como quieras, pero `i` es el estándar por "índice").
    * `{1..5}`: Es el rango. Bash entiende que debe contar del 1 al 5.
2.  **`do`**: Significa "Haz esto...". Es el inicio del bloque de repetición.
3.  **`$i`**: Dentro del bucle, usamos la variable para que el nombre de la carpeta cambie en cada vuelta (`Clase-1`, `Clase-2`, etc.).
4.  **`done`**: Significa "Terminado". Aquí el script vuelve arriba a ver si falta algún número o termina.



---

### ✍️ Nota para tu Cuaderno `.ipynb` (El truco del experto)

> **Rangos Flexibles:**
> En Bash puedes hacer rangos de números `{1..100}` o incluso de letras `{a..z}`. 
> También puedes usar listas personalizadas: `for color in rojo verde azul; do ... done`.
